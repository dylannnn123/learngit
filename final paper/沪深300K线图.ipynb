```python
from pyecharts import options as opts
from pyecharts.charts import Kline, Line,Bar
```

```python
from jqdatasdk import *
auth('18817917327','917327')
```

```python
date=get_trade_days(start_date='2018-01-01',end_date='2018-12-31')
date=date.tolist()
```

```python
price=get_price('000300.XSHG',start_date='2018-01-01',end_date='2018-12-31',frequency='daily')
```

```python
price['volume']=price['volume']/10000000
```

```python
price=price.drop(['money'],axis=1)
pricek=price.drop(['volume'],axis=1)
```

```python
train_data = np.array(price) #先将数据框转换为数组
price_list = train_data.tolist()  #其次转换为列表
```

```python
train_data = np.array(pricek) #先将数据框转换为数组
pricek_list = train_data.tolist()  #其次转换为列表
```

```python
price_list[0:3]
```

```python
pricek_list[0:3]
```

```python
kline=(
        Kline()
            .add_xaxis(date)
            .add_yaxis('日K线',pricek_list)
            .set_global_opts(
                xaxis_opts=opts.AxisOpts(is_scale=True),
                yaxis_opts=opts.AxisOpts(
                    is_scale=True,
                    splitarea_opts=opts.SplitAreaOpts(
                        is_show=True, areastyle_opts=opts.AreaStyleOpts(opacity=1)
                    ),
                ),
                datazoom_opts=[opts.DataZoomOpts()],
                title_opts=opts.TitleOpts(title="沪深300K线图（2018）")
            
            )
)
```

```python
kline.render_notebook()
```

```python
#这里只要计算两个移动平均
import pandas as pd
ma_list=[5,10]
for ma in ma_list:
    price['MA_'+str(ma)]=[round(i,2) for i in pd.Series(price['close']).rolling(ma).mean()]
    
print(price[0:10])
```

```python
line=(
        Line()
    .add_xaxis(date)
    .add_yaxis('ma_5',price['MA_5'].tolist(),is_smooth=True,label_opts=opts.LabelOpts(is_show=False))
    .add_yaxis('ma_10',price['MA_10'].tolist(),is_smooth=True)

)
line.render_notebook()
```

```python
kline.overlap(line)
```

```python
kline.render_notebook()
```

```python
bar=(
Bar().add_xaxis(date)
    .add_yaxis('volume',price['volume'].tolist(),label_opts=opts.LabelOpts(is_show=False))
    .set_global_opts(
        yaxis_opts=opts.AxisOpts(max_=7*max(price['volume']),is_show=False))
        
    )
bar.render_notebook()
```

```python
kline.overlap(bar)
kline.render_notebook()
```
